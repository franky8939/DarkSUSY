```python
from modules.general.F_search import *
from modules.graph.hist1D import *

import matplotlib.pyplot as plt
import numpy as np
import h5py
import sys
import re

# Create general h5 for all data
INPUT = "/home/franky8939/GITHUP/DarkSUSY-master/data/h5_muon_histg/DarkSUSY_histg.h5"

# Si existe el archivo OUTPUT ACTUALIZARLO #
if os.path.exists(INPUT):
    hf = h5py.File(INPUT, 'r')
else:
    print(" :: Datos de entrada inexistentes:: ")
    sys.exit()

# code to find
def find_Mu_Prop(vector, variable="Mu_Entries", file=hf):
    name_local_group = "MNeuL_" + str(vector[0]) + "/MNeuD_" + str(vector[1]) + \
                       "/MPhoD_" + str(vector[2]) + "/TcPhoD_" + str(vector[3]) + \
                       "/" + vector[4]
    # print(name_local_group)
    try:
        output = np.array(file.get(name_local_group + "/" + variable))
    except:
        output = None
    return output

# graficar histograma especifico para este caso
def graficar_plot(ax, vector, variable, color="red"):
    var = find_Mu_Prop(vector, variable=variable)
    ax.plot(var[0, :], var[1, :], alpha=0.7, label=vector[4]) # str(vector[4]))
    ax.set_title("parametros : MNeuL = " + str(vector[0]) + ", MNeuD = " + str(vector[1]) +
                 ", MPhoD = " + str(vector[2]) + ", TcPhoD = " + str(vector[3]))
    ax.grid(True)
    return ax
```

```python
variable = "IsolationVar"
plt.rcParams['figure.figsize'] = [30, 10]
fig = plt.figure()

ax = fig.add_subplot(2, 3, 1)
vector = [10, 0.25, 9, 0.5, "_CMS_"]
ax = graficar_plot(ax, vector, variable=variable)
ax.set_label("CMS")
vector = [10, 0.25, 9, 0.5, "_HL_"]
ax = graficar_plot(ax, vector, variable=variable)
ax.legend(loc='lower right', shadow=True)
# ax.set_xlim(vector[2]-1,vector[2]+1)

ax = fig.add_subplot(2, 3, 2)
vector = [10, 0.25, 9, 5, "_CMS_"]
ax = graficar_plot(ax, vector, variable=variable)
vector = [10, 0.25, 9, 5, "_HL_"]
ax = graficar_plot(ax, vector, variable=variable)
ax.legend(loc='lower right', shadow=True)
# ax.set_xlim(vector[2]-1,vector[2]+1)

ax = fig.add_subplot(2, 3, 3)
vector = [10, 0.25, 9, 10, "_CMS_"]
ax = graficar_plot(ax, vector, variable=variable)
vector = [10, 0.25, 9, 10, "_HL_"]
ax = graficar_plot(ax, vector, variable=variable)
ax.legend(loc='lower right', shadow=True)
#ax.set_xlim(vector[2]-1,vector[2]+1)

plt.show()
```
